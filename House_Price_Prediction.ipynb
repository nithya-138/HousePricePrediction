```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
mydata=sklearn.datasets.fetch_california_housing()
```

```python
print(mydata)
```

```python
mydata_dataframe=pd.DataFrame(mydata.data)
mydata_dataframe.head()
```

```python
mydata_dataframe=pd.DataFrame(mydata.data,columns=mydata.feature_names)
mydata_dataframe.head()
```

```python
mydata_dataframe['price']=mydata.target
```

```python
mydata_dataframe.head()
```

```python
mydata_dataframe.shape
```

```python
mydata_dataframe.isnull() 
```

```python
mydata_dataframe.isnull().sum() 
```

```python
mydata_dataframe.describe()
```

```python
correlation=mydata_dataframe.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```

```python
mydata_X=mydata_dataframe.drop(['price'],axis=1)
mydata_Y=mydata_dataframe['price']
```

```python
mydata_X_train,mydata_X_test,mydata_Y_train,mydata_Y_test=train_test_split(mydata_X,mydata_Y,test_size=0.2,random_state=2)
```

```python
print(mydata_X.shape,mydata_X_train.shape,mydata_X_test.shape)
```

```python
model=XGBRegressor()
model.fit(mydata_X_train,mydata_Y_train)
```

```python
mydata_Y_predict=model.predict(mydata_X_test)
```

```python
print(mydata_Y_predict)
```

```python
score_1=metrics.r2_score(mydata_Y_test,mydata_Y_predict)
score_2=metrics.mean_absolute_error(mydata_Y_test,mydata_Y_predict)
```

```python
print(score_1)
print(score_2)
```

```python
plt.scatter(mydata_Y_test,mydata_Y_predict)
plt.xlabel('Actual price')
plt.ylabel('Predicted price')
plt.title('Actual price vs Predicted price')
plt.show()
```
